#Load packages:

```
import numpy as np
import matplotlib.pyplot as plt

def mysqrt(x):
    if np.any(np.angle(x) > 0): 
        return np.exp(1j * np.angle(x) / 2) * np.sqrt(np.sqrt(np.real(x)**2 + np.imag(x)**2))
    else:
        return np.exp(1j * (np.angle(x)+2*np.pi) / 2) * np.sqrt(np.sqrt(np.real(x)**2 + np.imag(x)**2))
```

#Plot distributions

Semicircle distribution:

```
plt.style.use('seaborn-darkgrid')
fig, axs = plt.subplots(ncols=1, figsize=(12, 6))

t = np.linspace(0, 1, 10000)
t = 6*(t-0.5)
y = np.zeros(len(t))
y_l = 0
y_r = 0
y2 = np.zeros(len(t))
y2_l = 0
y2_r = 0
y3 = np.zeros(len(t))
y3_l = 0
y3_r = 0
for j in range(0, len(t)):
    if 4-t[j]**2 < 0:
        y[j] = 0
    else:
        y[j] = np.sqrt(4-t[j]**2)/(2*np.pi)

    if 4/4-t[j]**2 < 0:
        y2[j] = 0
    else:
        y2[j] = np.sqrt(4/4-t[j]**2)/(2*np.pi/4)

    if 4/16-t[j]**2 < 0:
        y3[j] = 0
    else:
        y3[j] = np.sqrt(4/16-t[j]**2)/(2*np.pi/16)
for j in range(0, len(t)):
    if y[j] > 0:
        y_l = j-1
        break
for j in range(0, len(t)):
    if y[len(t)-j-1] > 0:
        y_r = len(t)-j+1
        break
for j in range(0, len(t)):
    if y2[j] > 0:
        y2_l = j-1
        break
for j in range(0, len(t)):
    if y2[len(t)-j-1] > 0:
        y2_r = len(t)-j+1
        break
for j in range(0, len(t)):
    if y3[j] > 0:
        y3_l = j-1
        break
for j in range(0, len(t)):
    if y3[len(t)-j-1] > 0:
        y3_r = len(t)-j+1
        break

axs.set_xlim(-3, 3)
axs.set_ylim(0, 1.5)
axs.set_aspect('equal') 
axs.set_aspect(1)
axs.plot(t[y3_l:y3_r], y3[y3_l:y3_r], lw=2, color="green", label = "$\sigma =0.25$")
axs.plot(t[y2_l:y2_r], y2[y2_l:y2_r], lw=2, color="blue", label = "$\sigma =0.5$")
axs.plot(t[y_l:y_r], y[y_l:y_r], lw=2, color="red", label = "$\sigma = 1$")
#axs.set_facecolor((1, 0.65, 0, 0.3))


axs.legend(loc='best')
fig.tight_layout()
plt.show()
```

F-transform of the semicircle distribution (conformal mapping):

```
t = np.linspace(-2, 2, 1000)

result = np.zeros(len(t), dtype=complex)
for j in range(0, len(result)):
    z = t[j] + 1j * 0.0001
    result[j] = 2 / (z-mysqrt(z*z-4))

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots(ncols=1, figsize=(10, 5))

#ax.set_title('$\kappa = $' + str(kappa))
plt.ylim(0, 1.5)
plt.xlim(-2, 2)
ax.set_aspect('equal', 'datalim')
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
ax.plot(result.real, result.imag, lw=2)
for k in range(0,100):
    result2 = k/100*result
    ax.plot(result2.real, result2.imag, lw=2, color=[0,0,1,0.1])
```

F-transform of the semicircle distribution (+ preimage of the upper half-plane):

```
number = 10

t = np.linspace(-2, 2, 1000)

result = np.zeros(len(t), dtype=complex)
for j in range(0, len(result)):
    z = t[j] + 1j * 0.0001
    result[j] = 2 / (z-mysqrt(z*z-4))

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots(ncols=1, figsize=(10, 5))

#ax.set_title('$\kappa = $' + str(kappa))
plt.ylim(0, 1.5)
plt.xlim(-2, 2)
ax.set_aspect('equal', 'datalim')
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
ax.plot(result.real, result.imag, lw=2)
for k in range(3,number):
    result2 = k/number*result
    ax.plot(result2.real, result2.imag, lw=2, color=[1,0,0,0.7])

t = np.linspace(-1, 1, 1000)

result = np.zeros(len(t), dtype=complex)
for j in range(0, len(result)):
    w = t[j] + 1j * np.sqrt(1-t[j]**2)
    result[j] = (w**2+1)/w

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots(ncols=1, figsize=(10, 5))

#ax.set_title('$\kappa = $' + str(kappa))
plt.ylim(-4, 4)
plt.xlim(-4, 4)
ax.set_aspect('equal', 'datalim')
plt.ylim(-4, 4)
plt.xlim(-4, 4)
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
ax.plot(result.real, result.imag, lw=2,color=(0,0,1))
for k in range(3,number):
  for j in range(0, len(result)):
    w = k/number*(t[j] + 1j * np.sqrt(1-t[j]**2))
    result[j] = (w**2+1)/w
  plt.ylim(-7, 7)
  plt.xlim(-7, 7)
  ax.plot(result.real, result.imag, lw=2, color=[1,0,0,0.7])
```

Arcsine distribution:

```
plt.style.use('seaborn-darkgrid')
fig, axs = plt.subplots(ncols=1, figsize=(12, 6))


t = np.linspace(0, 1, 10000)
t = 6*(t-0.5)
y = np.zeros(len(t))
y_l = 0
y_r = 0
y2 = np.zeros(len(t))
y2_l = 0
y2_r = 0
y3 = np.zeros(len(t))
y3_l = 0
y3_r = 0
for j in range(0, len(t)):
    if np.sqrt(2)-t[j] < 0 or t[j] + np.sqrt(2)< 0:
        y[j] = 0
    else:
        y[j] = 1/(np.sqrt((np.sqrt(2)-t[j])*(np.sqrt(2)+t[j]))*np.pi)
   
    if 4/4-t[j]**2 < 0:
        y2[j] = 0
    else:
        y2[j] = np.sqrt(4/4-t[j]**2)/(2*np.pi/4)

    if 4/16-t[j]**2 < 0:
        y3[j] = 0
    else:
        y3[j] = np.sqrt(4/16-t[j]**2)/(2*np.pi/16)
for j in range(0, len(t)):
    if y[j] > 0:
        y_l = j-1
        break
for j in range(0, len(t)):
    if y[len(t)-j-1] > 0:
        y_r = len(t)-j+1
        break
for j in range(0, len(t)):
    if y2[j] > 0:
        y2_l = j-1
        break
for j in range(0, len(t)):
    if y2[len(t)-j-1] > 0:
        y2_r = len(t)-j+1
        break
for j in range(0, len(t)):
    if y3[j] > 0:
        y3_l = j-1
        break
for j in range(0, len(t)):
    if y3[len(t)-j-1] > 0:
        y3_r = len(t)-j+1
        break

axs.set_aspect('equal')
axs.set_xlim(-2.5, 2.5)
axs.set_ylim(0, 2)
#axs.plot(t[y3_l:y3_r], y3[y3_l:y3_r], lw=2, color="green", label = "$\sigma =0.25$")
#axs.plot(t[y2_l:y2_r], y2[y2_l:y2_r], lw=2, color="blue", label = "$\sigma =0.5$")
axs.plot(t[y_l+1:y_r-1], y[y_l+1:y_r-1], lw=2, color="red", label = "$\sigma = 1$")
#axs.set_facecolor((1, 0.65, 0, 0.3))
axs.plot(t[y_l:y_l+2], y[y_l:y_l+2], lw=2, linestyle='dashed', color="red")
axs.plot(t[y_r-2:y_r], y[y_r-2:y_r], lw=2, linestyle='dashed', color="red")
#axs.plot(t[y2_l:y2_l+2], y[y2_l:y2_l+2], lw=2, linestyle='dashed', color="blue")
#axs.plot(t[y2_r-2:y2_r], y[y2_r-2:y2_r], lw=2, linestyle='dashed', color="blue")
#axs.plot(t[y3_l:y3_l+2], y[y3_l:y3_l+2], lw=2, linestyle='dashed', color="green")
#axs.plot(t[y3_r-2:y3_r], y[y3_r-2:y3_r], lw=2, linestyle='dashed', color="green")
axs.legend(loc='best')
fig.tight_layout()
plt.show()

```

F-transform of the arcsine distribution (conformal mapping):

```
t = np.linspace(-np.sqrt(2), np.sqrt(2), 1000)
result = np.zeros(len(t), dtype=complex)
for j in range(0, len(result)):
    z = t[j] + 1j * 0.000001
    result[j] = mysqrt(z*z-2)#2 / (z-mysqrt(z*z-4))

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots(ncols=1, figsize=(10, 5))

#ax.set_title('$\kappa = $' + str(kappa))
plt.ylim(0, 1.5)
plt.xlim(-2, 2)
ax.set_aspect('equal', 'datalim')
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
ax.plot(result.real, result.imag, lw=2)
#for k in range(0,100):
  #  result2 = k/100*result
  #  ax.plot(result2.real, result2.imag, lw=2, color=[0,0,1,0.1])
```

Normal distribution:

```
plt.style.use('seaborn-darkgrid')
fig, axs = plt.subplots(ncols=1, figsize=(12, 6))


t = np.linspace(0, 1, 10000)
t = 6*(t-0.5)
y = np.zeros(len(t))
y2 = np.zeros(len(t))
y3 = np.zeros(len(t))

for j in range(0, len(t)):
    y[j] = np.exp(-t[j]**2/2)/(np.sqrt(2*np.pi))
    y2[j] = np.exp(-(t[j]/2)**2/2)/(2*np.sqrt(2*np.pi))
    y3[j] = np.exp(-(t[j]/0.5)**2/2)/(0.5*np.sqrt(2*np.pi))

axs.set_aspect(aspect=1)
axs.set_xlim(-3, 3)
axs.set_ylim(0, 1.0)

axs.set_aspect(1)
axs.plot(t, y3, lw=2, color="green", label = "$\sigma = 0.5$")
axs.plot(t, y, lw=2, color="red", label = "$\sigma = 1$")
axs.plot(t, y2, lw=2, color="blue", label = "$\sigma = 2$")
#axs.set_facecolor((1, 0.65, 0, 0.3))
axs.legend(loc='best')
fig.tight_layout()
plt.show()
```
