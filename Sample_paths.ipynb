#Sample paths

Brownian motion:

```
N = 1000
time = np.linspace(0, 1, N)
dt = 1 /N
nB = 5

B = np.zeros([nB,N])

for k in range(1, N):
  for j in range(0, nB):
    B[j,k] = B[j,k-1] + np.random.normal(loc=0.0, scale=np.sqrt(dt), size=None)

plt.style.use('seaborn-darkgrid')
fig, axs = plt.subplots(ncols=1, figsize=(10, 5))

axs.set_title('Brownian motion')
for j in range(0, nB):
  axs.plot(time, B[j,:], lw=2)

plt.show()
```

Poisson process:

```
lamba = 200
T = np.random.exponential(scale = 1/lamba, size=300)
for j in range(1,300):
  T[j] = T[j] + T[j-1]
T
```
